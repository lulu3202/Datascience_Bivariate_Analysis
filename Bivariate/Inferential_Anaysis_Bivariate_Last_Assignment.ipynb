# Test the similarity between the degree_t(Sci&Tech) and specialisation(Mkt&HR) with respect to salary at significance level of 5%.(Make decision using Hypothesis Testing) 

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

#perform two-way ANOVA
model = ols('salary ~ C(degree_t) * C(specialisation)', data=dataset).fit()
anova_table=sm.stats.anova_lm(model, typ=2)
print(anova_table)
```

# Convert the normal distribution to standard normal distribution for salary column

```python
#To convert to salary column from normal distribution to standard normal distribution 
# Calculate the mean and standard deviation of the salary column
mean_salary = dataset['salary'].mean()
std_salary = dataset['salary'].std()

# Standardize the salary column
dataset['salary_standardized'] = (dataset['salary'] - mean_salary) / std_salary

# Display the first few rows to verify
print(dataset[['salary', 'salary_standardized']].head())
```

# What is the probability Density Function of the salary range from 700000 to 900000?

```python
#calculating probability density 
def get_pdf_probability(dataset,startrange,endrange):
    from matplotlib import pyplot
    from scipy.stats import norm
    import seaborn as sns
    ax = sns.distplot(dataset,kde=True,kde_kws={'color':'blue'},color='Green')
    pyplot.axvline(startrange,color='Red')
    pyplot.axvline(endrange,color='Red')
    # generate a sample
    sample = dataset
    # calculate parameters
    sample_mean =sample.mean()
    sample_std = sample.std()
    print('Mean=%.3f, Standard Deviation=%.3f' % (sample_mean, sample_std))
    # define the distribution
    dist = norm(sample_mean, sample_std)
    
    # sample probabilities for a range of outcomes
    values = [value for value in range(startrange, endrange)]
    probabilities = [dist.pdf(value) for value in values]    
    prob=sum(probabilities)
    print("The area between range({},{}):{}".format(startrange,endrange,sum(probabilities)))
    return prob
```

```python
get_pdf_probability(dataset["salary"],700000,900000)
```

```python
#The computed probability or liklihood that a salary falls between 7 L and 9 L is very low 0.000597, or 0.0597%.
```
